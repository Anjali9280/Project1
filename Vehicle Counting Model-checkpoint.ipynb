```python
import cv2
import time
```

```python
# Importing the captured video for vehchile detection.
video_camera = cv2.VideoCapture("C:/Users/admin/Downloads/20201230123300.mp4")

#Reading the frame one by one from the video
ret, frame1 = video_camera.read()
# Initialize the counter of Vehchile as 0
vehchile=0
# Detecting and counting the passed vehchile till the videon is on.
while video_camera.isOpened():
    
    # Capture frame-by-frame and store in variable "Frame2"
    ret,frame2 = video_camera.read()  # capturing next frame
    frame= frame2.copy() # copy of frame2 is stored in another variable "frame"
    if not ret:
        break
        
    # Obtaining foreground image to detect object by removing background    
    foregroundMask = cv2.absdiff(frame1,frame2)
    
    #Converting the image into gray scale
    foregroundMask = cv2.cvtColor(foregroundMask, cv2.COLOR_BGR2GRAY)
    
    #Converting the gray scale image into Threshold
    _, thresh = cv2.threshold(foregroundMask, 50, 255, cv2.THRESH_BINARY)
    # Now frame 2 will become frame 1 till all the the frames are read 
    frame1= frame2
    
    # Creating a refrence line for counting the vehchile passed through
                    #(x1,y1),(x2,y2)     #color used in line
    cv2.line(frame,(400,450),(1300,450),(0,0,255),2) # Red Line
    cv2.line(frame,(400,440),(1300,440),(0,255,0),1) # Green Line Above the Red Line 
    cv2.line(frame,(400,460),(1300,460),(0,255,0),1) # Green line below the Red Line
    
    
    # Extracting the contours
    conts, _ = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE) 
        
    for c in conts:
        if cv2.contourArea(c) < 300:
            continue
            
        # Bounding the rectangles around     
        x,y,w,h = cv2.boundingRect(c)
        cv2.rectangle(frame,(x,y),(x+w, y+h),(0,255,0),2)
        xMid= int((x+(x+w))/2)
        yMid= int((y+(y+h))/2)
        cv2.circle(frame,(xMid,yMid),5,(0,0,255),5)
        
        
        # If vehchile would be passed between the two green lines then it will be counted as '1"
        if yMid > 440 and yMid < 460:
            vehchile +=1
            
     # Showing the output of the final detected video
    cv2.imshow("Foreground Mask",foregroundMask)
    cv2.putText(frame,"Total vehchile :{}".format(vehchile),(450,50),cv2.FONT_HERSHEY_PLAIN, 5, (255,0,0),2)
    cv2.imshow("Original Video",frame)
     
    if cv2.waitKey(1) & 0xFF== ord('q'):
        break

# When everything is done, release the capture
video_camera.release()
cv2.destroyAllWindows()
```
